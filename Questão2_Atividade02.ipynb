```python
import numpy as np
from statistics import mode
import matplotlib.pyplot as plt

TemposResposta = [32, 45, 28, 62, 54, 48, 37, 41, 29,
51,42,33,50,28,52,28,28]

media = np.mean(TemposResposta)
mediana = np.median(TemposResposta)
moda = mode(TemposResposta)

plt.bar(['media', 'mediana', 'moda'], [media, mediana, moda])
plt.title('media, mediana, moda')

print(f"\nMedia: {media:.2f}")
print(f"\nMediana: {mediana:.1f}")
print(f"\nModa: {moda:.2f}\n")
plt.show()
```
